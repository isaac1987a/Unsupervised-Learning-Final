```python
data_train = data_train[data_train['Text'].str.len() <= 7500]
print(data_train)
print(data_train.shape)
```

Now we will use Word2Vec to create word embeddings. Will use this to tokenize the text and eventually perform a sentiment analysis. I decided to use senitment analysis to see how the model rates a positive, negative, or neutral sentiment in articles.
Word2Vec is a two-layer neural net that processes text by “vectorizing” words. This will be useful when I perform NMF later on.

```python
from gensim.models import Word2Vec
import nltk
#nltk.download('punkt') Only need to do this once

from nltk.tokenize import word_tokenize, sent_tokenize

```

```python
# creating a dictionary to store tokenized text for each category
tokenized_text_by_category = {}

# looping through each unique category
for category in data_train['Category'].unique():
    #select the text content for the current category
    category_text = data_train[data_train['Category'] == category]['Text']
    
    #tokenize the text content using word_tokenize
    tokenized_text = [word_tokenize(article) for article in category_text]
    
    # Store the tokenized text in the dictionary
    tokenized_text_by_category[category] = tokenized_text

# can access the tokenized text for each category using the category labels
print(tokenized_text_by_category['politics'][0]) # Print tokenized text for the first article in the chosen category.
```

Now that we've tokenized our data, we are going to perform sentiment analysis on the tokenized text to retrieve a score on sentiment. 

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#nltk.download('vader_lexicon') #only need to download once 
analyzer = SentimentIntensityAnalyzer()
```

```python
# pefroming sentiment analysis
# creating a new column to store sentiment scores

# loop through each category
for category, tokenized_text_list in tokenized_text_by_category.items():
    # Perform sentiment analysis on each article's tokenized text
    sentiment_scores = [analyzer.polarity_scores(' '.join(article))['compound'] for article in tokenized_text_list]
    
    
    #storing the sentiment scores
    data_train.loc[data_train['Category'] == category, 'Sentiment_Scores'] = sentiment_scores
#function we can call to categorize sentiment scores.
def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05: #just choosing arbitrary values to define a positive, negative, or neutral
        return 'Negative'
    else:
        return 'Neutral'

data_train['Sentiment_Category'] = data_train['Sentiment_Scores'].apply(categorize_sentiment)
```

Now we will visiualize the sentiment scores and analysis

```python
#plot distribution of sentiment scores for each category
sns.boxplot(x='Category', y='Sentiment_Scores', data=data_train)
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='Category', hue='Sentiment_Category', data=data_train)
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```

From a quick glance, we can see that there are many positive sentiments across topics that I initially expected. It is not a surprise that politics and business were topics that recieved more negative sentiments. The model does not appear to show much sentinment with neutral ratings.

Unspervised Learning: Preparing and preprocessing data again to perfrom NMF.

```python
#calculate sentiment scores for each category
sentiment_scores_by_category = {} #dictionary

#looping through each category
for category, tokenized_text_list in tokenized_text_by_category.items():
    #performing sentiment analysis on each article's tokenized text
    sentiment_scores = [analyzer.polarity_scores(' '.join(article))['compound'] for article in tokenized_text_list]
    
    #storing the sentiment scores in the dictionary
    sentiment_scores_by_category[category] = sentiment_scores
```

Now we will perform NMF below.

```python
from sklearn.decomposition import NMF

#calculating the max number of words from each article
max_words = max([len(article) for category_text in tokenized_text_by_category.values() for article in category_text])

# Calculating the minimum sentiment score across all articles
min_sentiment = min([score for category_scores in sentiment_scores_by_category.values() for score in category_scores])

# Shift sentiment scores by the minimum value and add a constant value to make them non-negative.
# This is so that we can use NMF, as we cannot input negative values which I used to get a sentiment value.
shifted_sentiment_scores = {category: [score - min_sentiment + 0.1 for score in category_scores] for category, category_scores in sentiment_scores_by_category.items()}

#creating the rating matrix
rating_matrix = np.zeros((len(data_train), max_words))

#filling the rating matrix with sentiment scores
for idx, (category_text) in enumerate(tokenized_text_by_category.items()):
    #sentiment_scores = data_train[data_train['Category'] == category]['Sentiment_Scores']
    sentiment_scores = shifted_sentiment_scores[category]
    rating_matrix[idx, :len(sentiment_scores)] = sentiment_scores

#random hyperparameter
num_latent_factors = 10

#use NMF for matrix factorizaton
model = NMF(n_components=num_latent_factors, max_iter=7000, init='random', random_state=42)
user_features = model.fit_transform(rating_matrix)
item_features = model.components_

#predict the sentiment scores for train and test data
train_sent_predict = np.dot(user_features, item_features)
```

Now that we've performed NMF and have predicted the senitment scores for our train and test data, we will see how Logistic Regression performs.
We will also obtain accuracy and precision just to look at these metrics.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


y_train = data_train['Sentiment_Category']
X_train = train_sent_predict  # Sentiment analysis predictions

#training Logistic Regression Model
logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train, y_train)

#now we evaluate the Logistic Regression Model
# and obtain precision and recall.
y_pred = logreg_model.predict(X_train)
accuracy = accuracy_score(y_train, y_pred)
report = classification_report(y_train, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

Based off of our results with the Logistic Regression, it appears that our supervised model here correctly predicted 70% of the postive sentiment articles correctly. With a recall of 1.0, this means that all the postive sentiment articles were correctly labeled.
However, from the Classification report we see that the model did not perfrom well with the Neutral sentiment and struggled to classify this correctly. The model also didn't do as good fo a job with the negative classes either. I suspect this to be because I need to fine tune to model to correctly identify and classify what is a negative and/or neutral class.
